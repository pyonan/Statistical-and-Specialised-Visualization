```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from random import sample

np.random.seed(0)
df = pd.DataFrame({
    'Category': ['laptops', 'processors', 'graphics card', 'storage', 'ram'],
    'Value': sample(range(0,100),5),#[100, 150, 200, 205, 300],
    'Sales': sample(range(0,100),5),
    'Profit':sample(range(0,100),5)
})
display(df)

# Apply seaborn style
sns.set_style("darkgrid")

# For text and scalling
sns.set_context("paper")

# Set figure size
plt.figure(figsize=(8, 5))

# Area plot for Sales
plt.fill_between(df['Category'], df['Sales'],color='red', alpha=0.5, label='Sales',linewidth=2)

# Area plot for Profit
plt.fill_between(df['Category'], df['Profit'],color='green',alpha=0.25, label='Profit',linewidth=2)


# Titles and labels
plt.title("Sales and Profit Area Plot",fontweight='bold')
plt.xlabel("Category")
plt.ylabel("Values")
plt.legend()

# Show plot
plt.show()

# Pie Chart
# colors=["gold",'lime','salmon','navy','indigo']
# sns.set_style("darkgrid")
# sns.color_palette("pastel")

plt.figure()
plt.pie(df['Value'], labels=df['Category'], autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()

# If we want to remove one pie
# Explode only one slice (e.g., 3rd slice)
explode = [0.05, 0.05, 0.05, 0.05, 0.05]   # 0.1 = slight separation

plt.figure()
plt.pie(df['Value'], labels=df['Category'], autopct='%1.1f%%', explode=explode)
plt.title('Pie Chart')
plt.show()

# Donut Chart
explode = [0.05, 0.05, 0.05, 0.05, 0.05]
plt.figure()
plt.pie(df['Value'], labels=df['Category'], autopct='%1.1f%%')
centre_circle = plt.Circle((0, 0), 0.50, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Donut Chart')
plt.show()

# Box plot
plt.figure()
sns.boxplot(x=df['Value'],color='lightgreen')
plt.title("Boxplot of Sales")
plt.grid()
plt.show()

# Heatmap (Correlation)
plt.figure()
corr = df[['Sales', 'Profit','Value']].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()

# Bubble Plot
import matplotlib.pyplot as plt

plt.scatter(df['Sales'], df['Profit'],s=df['Value']*10,color='purple')   # size of bubbles

plt.title("Bubble Plot")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.show()

# Multi-colored bubble plot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.scatterplot(x='Sales', y='Profit',size='Value',hue='Value',data=df,sizes=(50, 300),palette='muted')
plt.title("Bubble Plot using Seaborn")
plt.show()

# Second Dataset
n = 100
df = pd.DataFrame({
    'Customer_ID': range(1, n+1),
    'Age': np.random.randint(20, 60, n),
    'Income': np.random.randint(20000, 100000, n),
    'Loan_Amount': np.random.randint(5000, 50000, n),
    'Credit_Score': np.random.randint(300, 900, n)
})

df['Loan_Status'] = np.where(df['Credit_Score'] > 600, 'Approved', 'Rejected')

print(df.head())

#Boxplot (Outlier Detection – Loan Amount)
import matplotlib.pyplot as plt

plt.boxplot(df['Loan_Amount'])

plt.title("Loan Amount Distribution (Boxplot)")
plt.ylabel("Loan Amount")

plt.show()

#Bubble Plot (Advanced Scatter)
import matplotlib.pyplot as plt
plt.scatter(df['Income'], df['Loan_Amount'],
            s=df['Credit_Score'],
            c=df['Credit_Score'],
            cmap='viridis',
            alpha=0.6)

plt.colorbar(label='Credit Score')

plt.title("Bubble Plot (Income vs Loan)")
plt.xlabel("Income")
plt.ylabel("Loan Amount")

plt.show()

#6. Correlation Heatmap (Very Important)
import seaborn as sns

corr = df[['Age','Income','Loan_Amount','Credit_Score']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("Correlation Matrix")

plt.show()
```
